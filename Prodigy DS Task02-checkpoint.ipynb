## Handling Missing Values

```python
df.isnull().sum()
```

```python
# Visualize missing values
sns.heatmap(df.isnull(),cbar=False)
plt.show()
```

```python
# fill with median or mode
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Drop 'Cabin' due to too many missing values
df.drop(columns=['Cabin'],inplace=True)
```

## Feature Relationships
#### Explore relationships between features and the target variable

```python
# If you want to include categorical data like Sex and Embarked in your correlation analysis, 
# you can convert them to numeric data:
# Convert categorical data to numeric using one-hot encoding
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64', 'uint8'])  # 'uint8' for one-hot encoded columns

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

## Handle Outliers

```python
# Boxplot to identify outliers
sns.boxplot(x=df['Fare'])
plt.show()
```

```python
# Handling outliers (example: capping the outliers)
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['Fare'] = df['Fare'].apply(lambda x: upper_bound if x > upper_bound 
                              else lower_bound if x < lower_bound else x)
```

## Feature Engineering
#### Create new features if needed

```python
# Example: Creating a new feature 'FamilySize'
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Creating a new feature 'IsAlone'
df['IsAlone'] = df['FamilySize'].apply(lambda x: 1 if x == 1 else 0)
```
